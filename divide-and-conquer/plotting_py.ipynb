# MergeSort vs QuickSort

```python
import matplotlib.pyplot as plt
import numpy as np

# Read data from file
sizes_m, times_m = np.loadtxt("dataMergeSort.txt", unpack=True)
sizes_q, times_q = np.loadtxt("dataQuickSort.txt", unpack=True)

# times = times / (sizes * np.log(sizes))

# Plot the data
plt.figure(figsize=(8,5))
plt.plot(sizes_m, times_m, marker='o', linestyle='-', color='b', label="MergeSort")
plt.plot(sizes_q, times_q, marker='o', linestyle='-', color='r', label="QuickSort")

# Set log scale for better visualization
plt.xscale('log')
plt.yscale('log')

plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (microseconds)")
plt.title("")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

plt.show()
```

```python
# compute slope
from scipy.stats import linregress

print(f'Slope quick sort: {linregress(np.log(sizes_q), np.log(times_q)).slope:.2f}')
print(f'Slope merge sort: {linregress(np.log(sizes_m), np.log(times_m)).slope:.2f}')
```
