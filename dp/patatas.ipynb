```python
G = 15
P = [2, 7, 4, 2, 1, 2, 1, 1, 3, 5, 6, 7, 2, 3, 4, 5, 10]
```

```python
def score(G):
    if G == 0:
        return float("inf")
    if G > 0:
        return -float("inf")
    return G

def patatas(G, P, j, memo):
    if (G,j) in memo:
        return memo[(G,j)]
    if j == 0:
        s = score(G)
    else:
        s = max(patatas(G - P[j - 1], P, j - 1, memo), patatas(G, P, j-1, memo))
    memo[(G,j)] = s
    return s
```

```python
memo = {}
patatas(G, P, len(P), memo)
```
